```python
sentence = "I am reading a book."

import nltk

# Download required resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
# Tokenization - breaking the sentence into small pieces like words
# Import tokenizer
from nltk.tokenize import word_tokenize

# Tokenize the sentence
tokens = word_tokenize(sentence)
tokens
```

```python
# 2. Stemming - to find the root word
from nltk.stem import PorterStemmer
# Create a stemmer object
stemmer = PorterStemmer()
# Apply stemming to each word
stemmed_words = [stemmer.stem(word) for word in tokens]
stemmed_words
```

```python
# 3.Lemmetazition - converts word into base or dictionary form (lemma)
nltk.download('wordnet')# required for lemmatization
from nltk.stem import WordNetLemmatizer

# Create lemmatizer object
lemmatizer = WordNetLemmatizer()

# Apply lemmatization (default POS = noun)
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]

lemmatized_words
```

```python
# 4.Remove the stop words
# stopwords - common words used in language that does not holds that much importance
nltk.download('stopwords')

from nltk.corpus import stopwords
# Load English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in tokens if word.lower() not in stop_words]

filtered_words
```
