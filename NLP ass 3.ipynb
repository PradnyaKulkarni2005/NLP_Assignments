```python
# Import required library
from collections import defaultdict

# Function to preprocess text
def preprocess(text):
    """
    Convert text to lowercase and split into words
    """
    text = text.lower()
    words = text.split()
    return words


# Function to create N-grams
def build_ngrams(words, n):
    """
    Create n-grams from the list of words
    Example: n=2 -> bigrams
    """
    ngrams = []

    for i in range(len(words) - n + 1):
        ngram = tuple(words[i:i+n])
        ngrams.append(ngram)

    return ngrams


# Function to train the N-gram model
def train_ngram_model(words, n):
    """
    Build frequency counts for n-grams
    """
    model = defaultdict(lambda: defaultdict(int))

    for i in range(len(words) - n + 1):
        context = tuple(words[i:i+n-1])   # previous words
        next_word = words[i+n-1]          # predicted word

        model[context][next_word] += 1

    return model


# Function to predict the next word
def predict_next_word(model, context):
    """
    Predict the most probable next word
    """
    context = tuple(context)

    if context not in model:
        return "No prediction available"

    # Get word with highest frequency
    next_words = model[context]
    predicted = max(next_words, key=next_words.get)

    return predicted


```

```python
# text data
text = """
Machine learning is fun and machine learning is powerful.
Learning models can predict words and learning improves models.
"""

# Step 1: preprocess text
words = preprocess(text)

# Step 2: choose N (2 = bigram model)
n = 2

# Step 3: train the model
model = train_ngram_model(words, n)

# Step 4: give context word
context = ["machine"]

# Step 5: predict next word
prediction = predict_next_word(model, context)

print("Context:", context)
print("Predicted next word:", prediction)
```
