```python
import nltk
# required libraries
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text="Singing is my hobby and I enjoy singing always."

# Tokenization
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text)
tokens
# POS Tagging
pos_tags=nltk.pos_tag(tokens)
print(pos_tags)
```
